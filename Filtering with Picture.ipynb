```sos
#Inserting the wanted libraries
%matplotlib notebook
%matplotlib inline
from ipywidgets import interact, widgets
from IPython.display import display
import matplotlib.pyplot as plt
import warnings
from skimage.morphology import disk
from skimage.filters import rank
import skimage.color as skic
import skimage.data
```

```sos
#Defining a function with Median filtering to experiment on a single image.
def f(Median_Size):
    selem = disk(int(Median_Size))
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        img_median = rank.median(img_gray, selem=selem) 

    ax_neu.imshow(img_median, cmap="gray")
    fig.canvas.draw()
    display(fig)
```

```sos
#Implementing an image and plotting
image = skimage.data.camera()      
img_gray = skic.rgb2gray(image)

fig = plt.figure(figsize=(6, 4))
ax_orig = fig.add_subplot(121) 
ax_neu = fig.add_subplot(122) 

ax_orig.imshow(img_gray, cmap="gray")
ax_neu.imshow(img_gray, cmap="gray")
```

```sos
#Addting a slider to adjust the noise. 
interact(f, Median_Size=widgets.IntSlider(min=1,max=21,step=2,value=1));
```
